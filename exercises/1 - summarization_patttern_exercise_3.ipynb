```python
import pandas as pd
```

```python
# Datasets' Directory

forum_node_directory = 'data_sets/forum_node.tsv'
forum_users_directory = 'data_sets/forum_users.tsv'

forum_node_results = forum_users_results = None
```

----

<h3 style="color:blue">Processing the <i>forum_node.tsv</i> dataset</h3>

```python
# Reading the forum_node dataset

df_forum_node = pd.read_csv(forum_node_directory, sep='\t')
df_forum_node
```

```python
# Filtering the columns
#
# \ id
# \ title
# \ tagnames
# \ author_id
# \ node_type
# \ parent_id
# \ abs_parent_id
# \ added_at

df_forum_node = df_forum_node.drop(df_forum_node.iloc[:, 4:5], axis=1)
df_forum_node = df_forum_node.drop(df_forum_node.iloc[:, 9:], axis=1)
```

```python
df_forum_node
```

----

```python
# Filling NaN values with '---'
df_forum_node.isnull().sum()
```

```python
df_forum_node.fillna('---', inplace=True)
```

```python
df_forum_node.isnull().sum()
```

----

```python
# Filling \N values with '---'
df_forum_node
```

```python
# Function to replace \N values with '---'
replace_n_values = lambda value: '---' if value == '\\N' else value
```

```python
# Replacing \N values

df_forum_node['id']               =   df_forum_node['id'].apply(replace_n_values)
df_forum_node['title']            =   df_forum_node['title'].apply(replace_n_values)
df_forum_node['tagnames']         =   df_forum_node['tagnames'].apply(replace_n_values)
df_forum_node['author_id']        =   df_forum_node['author_id'].apply(replace_n_values)
df_forum_node['node_type']        =   df_forum_node['node_type'].apply(replace_n_values)
df_forum_node['parent_id']        =   df_forum_node['parent_id'].apply(replace_n_values)
df_forum_node['abs_parent_id']    =   df_forum_node['abs_parent_id'].apply(replace_n_values)
df_forum_node['added_at']         =   df_forum_node['added_at'].apply(replace_n_values)
df_forum_node['score']            =   df_forum_node['score'].apply(replace_n_values)
```

```python
df_forum_node
```

----

<h3 style="color:blue">Processing the <i>forum_users.tsv</i> dataset</h3>

```python
# Reading the forum_users dataset

df_forum_users = pd.read_csv(forum_users_directory, sep='\t')
df_forum_users
```

```python
# Checking out the Null values
df_forum_users.isnull().sum()
```

----

<h3 style="color:blue">Creating the Search Function</h3>

```python
df_forum_node
```

```python
df_forum_users
```

----

```python
# Capturing by input which index will be returned
index_to_return = 0

try: index_to_return = int(input('Type an index: '))
except ValueError: print('An integer was not informed. The default index will be returned. (index == 0)')
except: print('Error to capture the input. The default index will be returned. (index == 0)')
```

```python
# Searching the Node and User infos
#
# the 'squeeze' method transforms the dataframe (or the row of the dataframe in this situation)
# into a Series

forum_node_results = dict(df_forum_node.iloc[index_to_return])
forum_users_results = dict(df_forum_users.loc[df_forum_users['user_ptr_id'] \
                                              == forum_node_results['author_id']].squeeze())
```

```python
forum_node_results
```

```python
forum_users_results
```

----

```python
# Returning the Post Results

print('***Post Infos***')
print()
print('Id: '               ,   forum_node_results['id'])
print('Title: '            ,   forum_node_results['title'])
print('Tagnames: '         ,   forum_node_results['tagnames'])
print('Node Type: '        ,   forum_node_results['node_type'])
print('Parent Id: '        ,   forum_node_results['parent_id'])
print('Abs Parent Id: '    ,   forum_node_results['abs_parent_id'])
print('Added At: '         ,   forum_node_results['added_at'])
print('Score: '            ,   forum_node_results['score'])
```

```python
# Returning the User Results

print('***User Infos***')
print()
print('Id: ', forum_users_results['user_ptr_id'])
print('Reputation: ', forum_users_results['reputation'])
print('Gold: ', forum_users_results['gold'])
print('Silver: ', forum_users_results['silver'])
print('Bronze: ', forum_users_results['bronze'])
```
