```python
import pandas as pd
```

```python
# Key Words
keyword_1 = 'fantastic'
keyword_2 = 'fantastically'
rows_indexes_keyword_2 = []

# DataSet Directory
dataset_dir = 'data_sets/forum_node.tsv'
```

----

```python
# Reading the file of posts
data = pd.read_csv(dataset_dir, sep='\t')
data
```

```python
# Dropping all of the columns, but "id", "title" and "body"
# and transforming "title" and "body" column to lowercase

data = data.drop(data.iloc[:, 2:4], axis=1)
data = data.drop(data.iloc[:, 3:], axis=1)

data['title'] = data['title'].str.lower()
data['body'] = data['body'].str.lower()
```

```python
# Filling NaN values with '---' String and merging title and body columns
data.fillna('---', inplace=True)

data['title - body'] = data['title'] + " --- " + data['body']
```

```python
# Overview
data
```

----

```python
# Filtering the rows that has the keyword 1 ('fantastic') in 'title - body' column
title_body_keyword_1 = len(data[data['title - body'].str.contains(keyword_1)])

print('Number of Rows in Title - Body Column with Fantastic Keyword: ', title_body_keyword_1)
```

----

```python
# Filtering the rows that has the keyword 2 (fantastically) in 'title - body' column
# and getting the node id of the filtered rows

title_body_keyword_2 = len(data[data['title - body'].str.contains(keyword_2)])
nodes_id_keyword_2 = list(data[data['title - body'].str.contains(keyword_2)]['id'])
nodes_id_keyword_2.sort()

print('Number of Rows in Title - Body Column with Fantastically Keyword: ', title_body_keyword_2)
print('Rows Indexes in Title - Body Column with Fantastically Keyword: ', nodes_id_keyword_2)
```
